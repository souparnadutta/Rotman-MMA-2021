```python
df1 = pd.read_csv("dataset_Fall 2021.csv")
```

```python
df1.shape
```

```python
na_values = round(df1.isnull().sum() / len(df1.index) * 100.00,2)
na_values
```

```python
plt.figure(figsize= (20,4),dpi=300)
na_values[na_values>0].plot(kind = 'bar')
plt.title (' columns having null values')
plt.ylabel('% null values')
plt.show()
```
